# 5. Visualisasi Dataset dengan menggunakan Grafik Box Plot

```python
# Membuat box plot Distribusi Kualitas Tidur Berdasarkan Tingkat Aktivitas Fisik.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Physical Activity Level', y='Quality of Sleep', data=df)
plt.title('Distribusi Quality of Sleep Berdasarkan Physical Activity Level')
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.show()
```

Penjelasan:
- plt.figure(figsize=(10, 6)) :  Mengatur ukuran gambar Box Plot menjadi 10 x 8 inci.lebar dan 6 inci
- sns.boxplot(x='Physical Activity Level', y='Quality of Sleep', data=df): Membuat box plot dengan parameter x='Physical Activity Level' menunjukkan bahwa sumbu x akan merepresentasikan tingkat aktivitas fisik, sedangkan y='Quality of Sleep' menunjukkan bahwa sumbu y akan merepresentasikan kualitas tidur. Data yang digunakan berasal dari DataFrame df
- plt.title : Memberi judul pada plot
- plt.xlabel('Tingkat Aktivitas Fisik') : Memberi label untuk sumbu x sebagai tingkat aktivitas fisik
- plt.ylabel('Kualitas Tidur') : Memberi label untuk sumbu y sebagai kualitas tidur
- plt.show() : Menampilkan Box Plot

# 6. Menampilkan Korelasi Dataset menggunakan heatmap

```python
# Menghapus Kolom 'Person ID'
numeric_columns_no_id = df[numeric_columns].drop(columns=['Person ID'])

# Menghitung matrik korelasi untuk kolom dengan tipe data numerik
correlation_matrix_no_id = numeric_columns_no_id.corr()

#  Membuat heatmap dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_no_id, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Dataset "Sleep Health and Lifestyle"', fontsize=12)
plt.show()
```

Penjelasan:
- correlation_matrix_no_id = numeric_columns_no_id.corr() : Menghitung matriks korelasi untuk kolom dengan tipe data numerik, yaitu mengukur sejauh mana variabel-variabel dalam dataset berkorelasi satu sama lain, dengan nilai korelasi yang berkisar dari -1 hingga 1.
- plt.figure(figsize=(10, 8)) : Mengatur ukuran gambar heatmap menjadi 10 x 8 inci.
- sns.heatmap(correlation_matrix_no_id, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5) : Membuat heatmap dari matriks korelasi yang disebut correlation_matrix_no_id. Dengan parameter annot=True, artinya setiap nilai dalam heatmap akan ditampilkan sebagai anotasi, sedangkan cmap='coolwarm' untuk menentukan palet warna yang digunakan, yaitu transisi dari warna dingin ke hangat, sedangkan Parameter fmt='.2f' akan mengatur format angka yang ditampilkan dengan dua desimal, dan linewidths=0.5 menambahkan garis tipis di antara sel-sel heatmap, memberikan tampilan yang lebih terstruktur.
- plt.title('Heatmap Korelasi Dataset "Sleep Health and Lifestyle"', fontsize=12) : mengatur judul heatmap dengan ukuran font 12
- plt.show() : menampilkan plot Heatmap

# 8. Hasil Analisis Visualisasi Data Dengan Menggunakan Heatmap

**Analisis Korelasi Berdasarkan Heatmap:**
- Korelasi Positif yang Kuat:
  - Sleep Duration vs Quality of Sleep: Korelasi yang sangat kuat (0.88), menunjukkan bahwa semakin lama durasi tidur seseorang, semakin baik kualitas tidurnya.
  - Daily Steps vs Physical Activity Level: Terdapat korelasi yang cukup kuat (0.77), yang logis karena semakin tinggi aktivitas fisik, biasanya diukur melalui jumlah langkah harian.
- Korelasi Negatif yang Kuat:
  - Quality of Sleep vs Stres Level: Korelasi sangat negatif (-0.90), menunjukkan bahwa semakin tinggi tingkat stres, kualitas tidur semakin buruk.
  - Sleep Duration vs Tingkat Stres: Korelasi negatif kuat (-0.81), menunjukkan bahwa durasi tidur yang lebih lama terkait dengan tingkat stres yang lebih rendah.
  -Quality of Sleep vs Heart Rate: Korelasi negatif signifikan (-0.66), yang mengindikasikan bahwa kualitas tidur yang baik mungkin berhubungan dengan denyut jantung yang lebih rendah (indikator kesehatan yang lebih baik).
- Korelasi Lain:
  - Tingkat Aktivitas Fisik vs Duration/Quality of Sleep: Korelasi rendah antara aktivitas fisik dengan kualitas tidur (0.19) dan durasi tidur (0.21). Artinya, tidak ada hubungan kuat antara seberapa aktif seseorang secara fisik dengan kualitas atau durasi tidurnya.
  - Stres Level vs. Heart Rate: Korelasi positif moderat (0.67), menunjukkan bahwa individu dengan tingkat stres yang lebih tinggi mungkin memiliki denyut jantung yang lebih tinggi.
